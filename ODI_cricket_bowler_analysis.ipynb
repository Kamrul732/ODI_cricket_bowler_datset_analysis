Q7:  Rename the column names 

```python
#rename the columns name
df = df.rename(columns={'Mat':'Match', 
                        'Inns':'Innings',
                        'Wkts' : 'Wickets',
                        'Ave': 'Average',
                        'Econ': 'Economy',
                        'SR': 'Strike_Rate',
                        4: 'Four',
                        5: 'Five'})

display(df.head())
```

```python
# splitting the 'Player' column to get the information about 'Country'
df[["Player_Name", "Country"]] = df['Player'].str.split("(", expand=True)
# remove the ")" from the 'Country' column
df['Country'] = df['Country'].str.replace(")", "")
# dropping the 'Player' columns
df = df.drop('Player', axis=1)
display(df.head())
```

```python
# rearrange the columns
new_col_sequence = ['Player_Name', 'Country', 'Span', 'Match', 'Innings', 'Balls', 'Runs', 'Wickets',
       'Average', 'Economy', 'Strike_Rate','Four', 'Five']
df = df[new_col_sequence]
display(df.head())
```

Q8: How many players played for ICC?

```python
def icc_check(x):
    if "ICC" in x:
        return "Yes"
    else:
        return "No"
```

```python
df['played_for_ICC'] = df['Country'].apply(icc_check)
#print(df['played_for_ICC'].value_counts())
display(df.head())
```

```python
print(f"Number of players played for ICC = {df['played_for_ICC'].value_counts()['Yes']}")
```

```python
df['Country'] = df['Country'].str.replace("ICC/", "")
df['Country'] = df['Country'].str.replace("/ICC", "")
df['Country'] = df['Country'].str.replace("Asia/", "")
df['Country'] = df['Country'].str.replace("Afr/", "")

display(df.head())
```

Q9: How many different countries are present in this dataset? 

```python
df['Country'].value_counts()
```

```python
print(f"Number of unique countries in the dataset= {len(df['Country'].unique())}")
```

```python
# splitting the 'Span' column based on the "-"
df[['start_year', 'end_year']] = df['Span'].str.split("-", expand=True)

# removing the "Span" column
df = df.drop("Span", axis=1)

display(df.head())
```

```python
df['start_year'] = df['start_year'].astype('int') 
df['end_year'] = df['end_year'].astype('int')

print(df.info())
```

```python
df['years_played'] = df['end_year'] - df['start_year']

df = df.drop(['start_year', "end_year"], axis=1)

display(df.head(10))
```
